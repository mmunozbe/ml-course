```python
filepath = os.path.join(mainpath, "customer-churn-model/Customer-Churn-Model.txt")
```

```python
df3 = open(filepath, "r")
cols = df3.readline().strip().split(",")
n_cols = len(cols)
```

```python
# read the variable file with a for loop and create a pandas dataframe
counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []

for line in df3:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter += 1

print("The dataset has {} rows and {} columns".format(counter, n_cols))

df3 = pd.DataFrame(main_dict)
df3.head()
```

# Ejercicio procesar la url de medallas

```python
import pandas as pd
import os, urllib3

medals_url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vRgDmJGDkg6EYP8aTOyOM2bU1Q8PBCi9HImsXr-MVKj08gJkG_c5OsfiOUCmYYLzUWa6-diTJXdu60K/pub?output=csv"
http = urllib3.PoolManager()
r = http.request('GET', medals_url)
response = r.data
# decode the response
response = response.decode('utf-8')
with open(os.path.join(mainpath, "athletes/downloaded_medals.csv"), "w") as file:
    file.write(response)

medalsdf = pd.read_csv(os.path.join(mainpath, "athletes/downloaded_medals.csv"))
print(medalsdf.head())
```
