# Data visualization

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("/home/matheo/Documents/ml-course-self/datasets/customer-churn-model/Customer-Churn-Model.txt")
```

```python
df.head()
```

### Scatter plot

```python
df.plot(kind="scatter", x="Day Mins", y="Day Charge")
```

```python
df.plot(kind="scatter", x="Night Mins", y="Night Charge")
```

```python
figure, axs = plt.subplots(2,2, sharey=True, sharex=True)
```

```python
df.plot(kind="scatter", x="Day Mins", y="Day Charge", ax=axs[0][0])
df.plot(kind="scatter", x="Night Mins", y="Night Charge", ax=axs[0][1])
df.plot(kind="scatter", x="Day Calls", y="Day Charge", ax=axs[1][0])
df.plot(kind="scatter", x="Night Calls", y="Night Charge", ax=axs[1][1])
figure
```

### Histograma

```python
k = int(np.ceil(1+np.log2(3333)))
df.hist(column="Day Calls", bins=k)
plt.xlabel("Numero de llamadas al dia")
plt.ylabel("Frecuencia")
plt.title("Histograma de numero de llamadas al dia")
plt.show()
```

```python
plt.hist(df["Day Calls"], bins=k)   
plt.xlabel("Numero de llamadas al dia")
plt.ylabel("Frecuencia")
plt.title("Histograma de numero de llamadas al dia")
plt.grid(visible=True)
plt.show()
```

```python
sns.set_theme()
sns.histplot(df["Day Calls"], bins=k, color="blue")
```

### Boxplot

```python
df.boxplot(column="Day Calls", grid=False)
plt.ylabel("Numero de llamadas al dia")
plt.title("Boxplot de las llamadas diarias")
plt.show()
```

```python
plt.boxplot(df["Day Calls"])
plt.ylabel("Numero de llamadas al dia")
plt.title("Boxplot de las llamadas diarias")
plt.show()
```

```python
sns.boxplot(y="Day Calls", data=df).set_title("Boxplot de las llamadas diarias")
plt.show()
```
