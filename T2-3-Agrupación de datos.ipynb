# Agregación de datos por categoría

```python
import pandas as pd
import numpy as np
```

```python
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
n = 500
gender_data = []
income_data = []

for _ in range(0, n):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
gender_data[:10]
```

```python
income_data[:10]
```

```python
height = np.random.normal(160, 30, size=n)
weight = np.random.normal(65, 25, size=n)
age = np.random.normal(30, 12, size=n)
income = np.random.normal(18000, 3500, size=n)
```

```python
data = pd.DataFrame(
    {
        "Gender" : gender_data,
        "Economic Status" : income_data,
        "Height" : height,
        "Weight" : weight,
        "Age" : age,
        "Income" : income,
    }   
)
```

```python
data.head()
```

```python
data["Age"] = data["Age"].astype(int)
```

## Agrupación de datos

```python
grouped_gender = data.groupby("Gender")
```

```python
grouped_gender.groups
```

```python
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
grouped_gender.get_group("Female")
```

```python
data.groupby(["Gender", "Economic Status"]).get_group(("Female", "Middle Class")).shape[0]
```

```python
double_group = data.groupby(["Gender", "Economic Status"])
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

## Operaciones sobre datos agrupados

```python
double_group.mean()
```

```python
double_group.size()
```

```python
grouped_income = double_group["Income"]
```

```python
grouped_income.describe()
```

```python
double_group.aggregate(
    {
        "Income" : "sum",
        "Age" : "mean",
        "Height" : "std",
    }
)
```

```python
double_group.aggregate(
    {
        "Age" : "mean",
        "Height" : lambda h: np.mean(h) / np.std(h),
    }
)
```

```python
double_group.aggregate(["sum", "mean", "std"])
```

```python
double_group.aggregate([lambda x: np.mean(x) / np.std(x)])
```
